```python
import pandas as pd
import numpy as np
import pandas_datareader.data as wb
import matplotlib.pyplot as plt
```

```python
stocks = ['AAPL','AMZN','MSFT','TSLA']
data=wb.DataReader(stocks,data_source='yahoo',start='12/01/2017')['Adj Close']
```

```python
data
```

```python
returns=data.pct_change()
returns.head()

```

```python
mean=returns.mean()
cov_mat=returns.cov()
cov_mat
```

```python
iterations=1000
simulation=np.zeros((iterations,3+len(stocks)))
simulation
```

```python
for i in range(iterations):
    weight= np.array(np.random.random(4))
    weight /= np.sum(weight)
    port_return=np.sum(returns.mean()*weight)*250
    port_std = np.sqrt(np.dot(weight.T,np.dot(cov_mat,weight)))*np.sqrt(250)
    simulation[i,0]=port_return
    simulation[i,1]=port_std
    simulation[i,2]=simulation[i,0]/ simulation[i,1]                        
    for j in range(len(weight)):
        simulation[i,j+3]=weight[j]



    
```

```python
sim_frame= pd.DataFrame(simulation,columns=['ret','std','sharpe',stocks[0],stocks[1],stocks[2],stocks[3]])
```
